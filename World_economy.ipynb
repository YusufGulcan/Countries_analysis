```python
import pandas as pd
import numpy as np 
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = r"C:\Users\yusuf\OneDrive\Documents\Data Science Projects\Database\Streamlit Project"
Agriculture_percent = "agriculture_percent.csv"

Ease_of_business = 'Ease_of_business.csv'
Exp_education = 'exp_education.csv'
Export = 'export.csv'
GDP_world = 'GDP_world.csv'
Health_exp = 'health_exp.csv'
Import_world = 'import.csv'
Industry_percent = 'Industry_percent.csv'
Inflation_world = 'Inflation_world.csv'
RnD = 'RnD.csv'
Service_percent = 'Service_percent.csv'
Unemployment = 'unemployment.csv'
Population = "Population.csv"

list_of_paths = [Agriculture_percent,Ease_of_business,Exp_education,Export,GDP_world,Health_exp,Import_world,Industry_percent,Inflation_world,RnD,Service_percent,Unemployment,Population]
```

```python
def melt(name):
    path_current = path +"\\" + name
    df = pd.read_csv(path_current)
    cols_to_melt = df.columns[-24:-1]
    melted = pd.melt(df,id_vars = ["Country Name", "Country Code"], value_vars= cols_to_melt,var_name = "Year",value_name= f'{name[:-4]}')
    return melted
```

```python
final_frame = pd.DataFrame()
number = 1
for value in list_of_paths:
    print(value)
    df_current = melt(value)
    if number == 1:
        final_frame = df_current
        number = number +1
    else:
        final_frame = pd.merge(left = final_frame,right= df_current, how = "outer", on = ["Country Name", "Year","Country Code"])
        number = number + 1
```

```python
ccodes = pd.read_csv(path+"\\"+"country_code.csv")
ccodes = ccodes.loc[:,["Continent_Name","Three_Letter_Country_Code"]].dropna()
final_frame = pd.merge(left = final_frame, right = ccodes, left_on="Country Code", right_on="Three_Letter_Country_Code", how = 'left').drop(columns = "Three_Letter_Country_Code").dropna(subset = ["Continent_Name"])
final_frame = final_frame.sort_values(["Country Name","Year"])
final_frame
```
