```python
import pandas as pd
import numpy as np 
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
df = pd.read_csv("World_economy.csv")
df.rename(columns = {"GDP_world":"GDP", 
                     "Inflation_world" : "Inflation",
                     "agriculture_percent": "Agriculture_Percentage",
                     "exp_education":"Expenditure_on_Education",
                     "health_exp":"Expenditure_on_Health",
                     "export":"Export",
                     "import":"Import",
                     "unemployment":"Unemployment"},inplace =True)
df["Net_trade"] = df.Export - df['Import']
df.head()
```

### GDP Race

```python
gdp_data = df[df.GDP.isna()==False].groupby(["Continent_Name","Year"])["GDP"].sum().reset_index()
px.line(gdp_data, x = "Year", y = "GDP", color = "Continent_Name", height = 700,markers=True)
```

```python
countries_data = df[(df["Continent_Name"] == "Asia")&(df.GDP.isna()==False)]

px.line(countries_data , x = "Year", y = "GDP", color = "Country Name",height = 700,
        markers=True)
```

```python
choro_data = df[df.GDP.isna()==False].sort_values(["Year"])
px.choropleth(choro_data , locations ="Country Code", color = "GDP", 
              range_color=[0,3000000000000], width = 1400, height = 900,
               color_continuous_scale="Viridis",animation_frame="Year" )
```

```python
metricc = "GDP"

whiskers_data = df[df[metricc].isna()==False]
whisker_data1 = whiskers_data.groupby("Year")[metricc].mean().reset_index()
whisker_data1

fig = px.box(whiskers_data,x = "Year", y = metricc, color = 'Year', hover_data="Country Name")
fig.add_traces(go.Scatter(x = whisker_data1["Year"], y = whisker_data1[metricc], mode="lines+markers"))

```

```python
px.violin(whisker_data1, x = 'Continent_Name', y = metricc, color = "Continent_Name")
```

```python
px.histogram(whiskers_data, x= "Continent_Name", y = metricc,color = "Continent_Name")
```

```python
px.pie(whiskers_data, names = "Continent_Name", values=metricc,height=400, width=600)
```
